```python
import pandas as pd

# 파일 경로
file_path = "./data/공고원본_POSTING_ID포함_정리결과.xlsx"

# normalized_freq 시트 불러오기
df = pd.read_excel(file_path, sheet_name="normalized_freq")

# tech_stack 정리
df["tech_stack"] = df["tech_stack"].astype(str).str.strip()

# -----------------------------
# 1. 카테고리 사전 정의
# -----------------------------
category_map = {
    # 백엔드
    "java": "백엔드",
    "spring": "백엔드",
    "spring boot": "백엔드",
    "jsp": "백엔드",
    "servlet": "백엔드",
    "node.js": "백엔드",
    "nodejs": "백엔드",
    "express": "백엔드",
    "django": "백엔드",
    "flask": "백엔드",
    "fastapi": "백엔드",
    "php": "백엔드",
    "laravel": "백엔드",
    "mybatis": "백엔드",
    "jpa": "백엔드",
    "hibernate": "백엔드",
    "oracle": "백엔드",
    "mysql": "백엔드",
    "mariadb": "백엔드",
    "postgresql": "백엔드",
    "postgres": "백엔드",
    "mssql": "백엔드",
    "redis": "백엔드",

    # 프론트
    "html": "프론트",
    "css": "프론트",
    "javascript": "프론트",
    "typescript": "프론트",
    "react": "프론트",
    "vue": "프론트",
    "angular": "프론트",
    "next.js": "프론트",
    "nextjs": "프론트",
    "nuxt.js": "프론트",
    "nuxtjs": "프론트",
    "jquery": "프론트",
    "react native": "프론트",
    "flutter": "프론트",
    "android": "프론트",
    "ios": "프론트",

    # 인공지능
    "python": "인공지능",
    "tensorflow": "인공지능",
    "pytorch": "인공지능",
    "keras": "인공지능",
    "scikit-learn": "인공지능",
    "sklearn": "인공지능",
    "pandas": "인공지능",
    "numpy": "인공지능",
    "matplotlib": "인공지능",
    "seaborn": "인공지능",
    "opencv": "인공지능",
    "nlp": "인공지능",
    "machine learning": "인공지능",
    "deep learning": "인공지능",
    "llm": "인공지능",
    "bert": "인공지능",
    "transformer": "인공지능",
    "langchain": "인공지능",
    "rag": "인공지능",

    # 인프라
    "aws": "인프라",
    "azure": "인프라",
    "gcp": "인프라",
    "docker": "인프라",
    "kubernetes": "인프라",
    "jenkins": "인프라",
    "linux": "인프라",
    "terraform": "인프라",
    "ansible": "인프라",
    "ci/cd": "인프라",
    "nginx": "인프라",
    "apache": "인프라",
    "github actions": "인프라",
    "devops": "인프라"
}

# -----------------------------
# 2. 애매한 키워드 정의
# -----------------------------
ambiguous_keywords = {
    "python", "sql", "git", "api", "rest api", "graphql",
    "kafka", "etl", "spark", "hadoop", "mongodb", "firebase"
}

# 기술스택으로 보기 애매한 도구/협업툴
tool_keywords = {
    "notion", "slack", "jira", "figma", "excel", "powerpoint", "word", "teams"
}

# -----------------------------
# 3. 정규화용 컬럼 생성
# -----------------------------
df["tech_stack_clean"] = df["tech_stack"].str.lower().str.strip()

# -----------------------------
# 4. 카테고리 분류 함수
# -----------------------------
def classify_category(stack):
    if stack in tool_keywords:
        return "분류보류"
    return category_map.get(stack, "분류보류")

def review_flag(stack):
    if stack in ambiguous_keywords:
        return "검토필요"
    if stack in tool_keywords:
        return "제외후보"
    return ""

def note_text(stack):
    if stack in ambiguous_keywords:
        return "문맥에 따라 여러 카테고리에 걸칠 수 있음"
    if stack in tool_keywords:
        return "기술스택보다 협업/업무도구 성격이 강함"
    return ""

# -----------------------------
# 5. 분류 적용
# -----------------------------
df["category"] = df["tech_stack_clean"].apply(classify_category)
df["review_flag"] = df["tech_stack_clean"].apply(review_flag)
df["note"] = df["tech_stack_clean"].apply(note_text)

# tech_stack_clean은 중간 작업용이니까 마지막엔 빼도 됨
result_df = df.drop(columns=["tech_stack_clean"])

# -----------------------------
# 6. 저장
# -----------------------------
output_path = "기술스택_카테고리분류_결과.xlsx"
result_df.to_excel(output_path, index=False)

print("저장 완료:", output_path)
print(result_df.head())
print("\n카테고리별 개수")
print(result_df["category"].value_counts())
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util
import torch

file_path = "./data/공고원본_POSTING_ID포함_정리결과.xlsx"
df = pd.read_excel(file_path, sheet_name="normalized_freq")

df["tech_stack"] = df["tech_stack"].astype(str).str.strip()

category_texts = {
    "백엔드": "server api database backend spring java node django flask fastapi sql jpa mybatis",
    "프론트": "frontend ui web react vue angular javascript typescript html css next",
    "인공지능": "artificial intelligence machine learning deep learning python tensorflow pytorch sklearn nlp llm data analysis",
    "인프라": "infrastructure aws docker kubernetes linux jenkins devops cloud deployment ci cd"
}

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

category_names = list(category_texts.keys())
category_descs = list(category_texts.values())

category_emb = model.encode(category_descs, convert_to_tensor=True)
keyword_emb = model.encode(df["tech_stack"].tolist(), convert_to_tensor=True)

sim_matrix = util.cos_sim(keyword_emb, category_emb)

categories = []
review_flags = []
scores = []

for sims in sim_matrix:
    sims_list = sims.tolist()
    sorted_idx = sorted(range(len(sims_list)), key=lambda i: sims_list[i], reverse=True)

    best_idx = sorted_idx[0]
    second_idx = sorted_idx[1]

    best_score = sims_list[best_idx]
    second_score = sims_list[second_idx]

    best_category = category_names[best_idx]

    # 기준은 나중에 조정 가능
    if best_score < 0.30:
        categories.append("분류보류")
        review_flags.append("")
    elif best_score - second_score < 0.05:
        categories.append(best_category)
        review_flags.append("검토필요")
    else:
        categories.append(best_category)
        review_flags.append("")

    scores.append(round(best_score, 4))

df["category"] = categories
df["review_flag"] = review_flags
df["score"] = scores

# 검토필요 + 분류보류만 따로
filtered_df = df[(df["review_flag"] == "검토필요") | (df["category"] == "분류보류")].copy()

with pd.ExcelWriter("기술스택_자동분류_결과.xlsx", engine="openpyxl") as writer:
    df.to_excel(writer, sheet_name="전체결과", index=False)
    filtered_df.to_excel(writer, sheet_name="검토필요_분류보류", index=False)

print("완료")
print(df["category"].value_counts())
```
